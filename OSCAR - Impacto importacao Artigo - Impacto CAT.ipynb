```python
# preparar a série de mapeamento (única por código)
map_series = (
    tradutor_atv_contas_nacionais
      .dropna(subset=['Cod_atv'])
      .assign(Cod_atv=lambda d: d['Cod_atv'].astype(str).str.strip())
      .drop_duplicates(subset='Cod_atv', keep='first')  # ou outra regra
      .set_index('Cod_atv')['Atividade de Divulgação - SCN 2010']
)

importacoes_mapeadas_scn['Descrição Atividade SCN'] = (
    importacoes_mapeadas_scn['Codigo Atividade SCN']
      .astype(str).str.strip()
      .map(map_series)
)
importacoes_mapeadas_scn.head(5)
```

```python
# Totais das importações e exportações do Mundo e dos EUA por Produto e por Conta do SCN
importacoes_mapeadas_scn['Total Importações Produto Mundo'] = importacoes_mapeadas_scn.groupby('Código Produto')['Importação - 2024 - Valor US$ FOB MUNDO'].transform('sum')
importacoes_mapeadas_scn['Total Importações SCN Mundo'] = importacoes_mapeadas_scn.groupby('Codigo Atividade SCN')['Importação - 2024 - Valor US$ FOB MUNDO'].transform('sum')
importacoes_mapeadas_scn['Total Importações Geral Mundo'] = sum(importacoes_mapeadas_scn['Importação - 2024 - Valor US$ FOB MUNDO'])

importacoes_mapeadas_scn['Total Importações Produto EUA'] = importacoes_mapeadas_scn.groupby('Código Produto')['Importação - 2024 - Valor US$ FOB EUA'].transform('sum')
importacoes_mapeadas_scn['Total Importações SCN EUA'] = importacoes_mapeadas_scn.groupby('Codigo Atividade SCN')['Importação - 2024 - Valor US$ FOB EUA'].transform('sum')
importacoes_mapeadas_scn['Total Importações Geral EUA'] = sum(importacoes_mapeadas_scn['Importação - 2024 - Valor US$ FOB EUA'])

importacoes_mapeadas_scn['Total Exportações Produto Mundo'] = importacoes_mapeadas_scn.groupby('Código Produto')['Exportação - 2024 - Valor US$ FOB MUNDO'].transform('sum')
importacoes_mapeadas_scn['Total Exportações SCN Mundo'] = importacoes_mapeadas_scn.groupby('Codigo Atividade SCN')['Exportação - 2024 - Valor US$ FOB MUNDO'].transform('sum')
importacoes_mapeadas_scn['Total Exportações Geral Mundo'] = sum(importacoes_mapeadas_scn['Exportação - 2024 - Valor US$ FOB MUNDO'])

importacoes_mapeadas_scn['Total Exportações Produto EUA'] = importacoes_mapeadas_scn.groupby('Código Produto')['Exportação - 2024 - Valor US$ FOB EUA'].transform('sum')
importacoes_mapeadas_scn['Total Exportações SCN EUA'] = importacoes_mapeadas_scn.groupby('Codigo Atividade SCN')['Exportação - 2024 - Valor US$ FOB EUA'].transform('sum')
importacoes_mapeadas_scn['Total Exportações Geral EUA'] = sum(importacoes_mapeadas_scn['Exportação - 2024 - Valor US$ FOB EUA'])


importacoes_mapeadas_scn.head(5)

```

```python
# Participação dos EUA Importações e Exportações do Mundo
importacoes_mapeadas_scn['Participação Importações EUA-Mundo por Produto'] = importacoes_mapeadas_scn['Total Importações Produto EUA'] / importacoes_mapeadas_scn['Total Importações Produto Mundo']
importacoes_mapeadas_scn['Participação Importações EUA-Mundo por SCN'] = importacoes_mapeadas_scn['Total Importações SCN EUA'] / importacoes_mapeadas_scn['Total Importações SCN Mundo']

importacoes_mapeadas_scn['Participação Exportações EUA-Mundo por Produto'] = importacoes_mapeadas_scn['Total Exportações Produto EUA'] / importacoes_mapeadas_scn['Total Exportações Produto Mundo']
importacoes_mapeadas_scn['Participação Exportações EUA-Mundo por SCN'] = importacoes_mapeadas_scn['Total Exportações SCN EUA'] / importacoes_mapeadas_scn['Total Exportações SCN Mundo']
```

```python
# Representatividade da NCM no Produto e do Produto na Conta SCN -- do mundo e dos EUA

importacoes_mapeadas_scn['Representatividade Importações NCM no Produto - Mundo'] = importacoes_mapeadas_scn['Importação - 2024 - Valor US$ FOB MUNDO'] / importacoes_mapeadas_scn['Total Importações Produto Mundo']
importacoes_mapeadas_scn['Representatividade Importações Produto na Conta SCN - Mundo'] = importacoes_mapeadas_scn['Total Importações Produto Mundo'] / importacoes_mapeadas_scn['Total Importações SCN Mundo']
importacoes_mapeadas_scn['Representatividade Importações da Conta SCN no Total - Mundo'] = importacoes_mapeadas_scn['Total Importações SCN Mundo'] / importacoes_mapeadas_scn['Total Importações Geral Mundo']

importacoes_mapeadas_scn['Representatividade Importações NCM no Produto - EUA'] = importacoes_mapeadas_scn['Importação - 2024 - Valor US$ FOB EUA'] / importacoes_mapeadas_scn['Total Importações Produto EUA']
importacoes_mapeadas_scn['Representatividade Importações Produto na Conta SCN - EUA'] = importacoes_mapeadas_scn['Total Importações Produto EUA'] / importacoes_mapeadas_scn['Total Importações SCN EUA']
importacoes_mapeadas_scn['Representatividade Importações da Conta SCN no Total - EUA'] = importacoes_mapeadas_scn['Total Importações SCN EUA'] / importacoes_mapeadas_scn['Total Importações Geral EUA']

importacoes_mapeadas_scn['Representatividade Exportações NCM no Produto - Mundo'] = importacoes_mapeadas_scn['Exportação - 2024 - Valor US$ FOB MUNDO'] / importacoes_mapeadas_scn['Total Exportações Produto Mundo']
importacoes_mapeadas_scn['Representatividade Exportações Produto na Conta SCN - Mundo'] = importacoes_mapeadas_scn['Total Exportações Produto Mundo'] / importacoes_mapeadas_scn['Total Exportações SCN Mundo']
importacoes_mapeadas_scn['Representatividade Exportações da Conta SCN no Total - Mundo'] = importacoes_mapeadas_scn['Total Exportações SCN Mundo'] / importacoes_mapeadas_scn['Total Exportações Geral Mundo']

importacoes_mapeadas_scn['Representatividade Exportações NCM no Produto - EUA'] = importacoes_mapeadas_scn['Exportação - 2024 - Valor US$ FOB EUA'] / importacoes_mapeadas_scn['Total Exportações Produto EUA']
importacoes_mapeadas_scn['Representatividade Exportações Produto na Conta SCN - EUA'] = importacoes_mapeadas_scn['Total Exportações Produto EUA'] / importacoes_mapeadas_scn['Total Exportações SCN EUA']
importacoes_mapeadas_scn['Representatividade Exportações da Conta SCN no Total - EUA'] = importacoes_mapeadas_scn['Total Exportações SCN EUA'] / importacoes_mapeadas_scn['Total Exportações Geral EUA']

```

## Calcular a matriz de coeficientes técnicos e demais componentes do modelo de preços

```python
matriz_guilhoto =pd.read_excel('./Outputs/MIP_2019_68x68MetodoGuilhoto.xlsx')
```

```python
matriz_guilhoto.tail()
```

```python
valor_producao = matriz_guilhoto.iloc[87,1:69]
```

```python
consumo_intermediario = matriz_guilhoto.iloc[:68,1:69]
```

```python
#matriz_guilhoto.iloc[69:74,0:69]
importacao_modelo_precos = matriz_guilhoto.iloc[69,1:69]
impostos_modelo_precos = matriz_guilhoto.iloc[70:74,1:69]

valor_agregado = matriz_guilhoto.iloc[75,1:69]
```

```python
coeficientes_tecnicos = consumo_intermediario.div(valor_producao, axis=1)
```

```python
#dividir cada linha da matriz de impostos_modelo_precos pelo valor de produção, de forma que o elemento de cada linha seja dividido pelo valor de produção, na coluna correspondente
importacao_modelo_precos_part = importacao_modelo_precos.div(valor_producao, axis=0)
valor_agregado_part = valor_agregado.div(valor_producao, axis=0)
# dividir cada linha da matriz de impostos_modelo_precos pelo valor de produção, de forma que o elemento de cada linha seja dividido pelo valor de produção, na coluna correspondente
impostos_modelo_precos_part = impostos_modelo_precos.div(valor_producao, axis=1)
```

```python
coeficientes_tecnicos.head()
```

```python
#concatenar os 4 vetores em um único DataFrame
matriz_modelo_precos = pd.concat([coeficientes_tecnicos,pd.DataFrame(importacao_modelo_precos_part.values.reshape(1,68),columns=coeficientes_tecnicos.columns),impostos_modelo_precos_part,pd.DataFrame(valor_agregado_part.values.reshape(1,68),columns=coeficientes_tecnicos.columns)], axis=0,ignore_index=True)
```

```python
coeficientes_tecnicos
```

```python
#Calcular a inversa de leontief, que é igual a (I - transposta(coeficientes_tecnicos))^-1

import numpy as np
from scipy.linalg import inv
I = np.eye(coeficientes_tecnicos.shape[1])
inversa_leontief = inv(I - coeficientes_tecnicos.values.T.astype(float))


```

```python
matriz_modelo_precos = pd.concat([pd.DataFrame(importacao_modelo_precos_part.values.reshape(1,68),columns=coeficientes_tecnicos.columns),impostos_modelo_precos_part,pd.DataFrame(valor_agregado_part.values.reshape(1,68),columns=coeficientes_tecnicos.columns)], axis=0,ignore_index=True)
```

```python
vetor_alteração_imposto
```

```python
#multiplicar cada elento do vetor de alteração pela segunda linha da matriz de modelo de preços, que é a linha de imposto de importação
matriz_modelo_precos.iloc[1,:] = matriz_modelo_precos.iloc[1,:] * vetor_alteração_imposto.values
```

```python
matriz_modelo_precos
```

```python
#somar todas as linhas coluna a coluna da matriz_modelo_precos
matriz_modelo_precos = matriz_modelo_precos.sum(axis=0)
```

```python
#multiplicar a inversa de leontief pela matriz de modelo de preços
resultado_modelo_precos = inversa_leontief @ matriz_modelo_precos.values
```
